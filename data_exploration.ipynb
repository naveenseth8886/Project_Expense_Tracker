```python
from sklearn.preprocessing import LabelEncoder

lb1= LabelEncoder()
lb2 = LabelEncoder()
```

```python
df['Occupation']= lb1.fit_transform(df['Occupation'])


df['City_Tier'] = lb2.fit_transform(df['City_Tier'])

```

```python
import pickle
pickle.dump(lb1, open('Health_care_model_City.pkl', 'wb'))
```

```python
pickle.dump(lb1, open('Health_care_model_Occupation.pkl', 'wb'))
```

```python
from sklearn.model_selection import train_test_split 
```

```python
df.head(1)
```

```python
x = df[['City_Tier' ,'Occupation','Dependents','Income','Age','Loan_Repayment','Insurance','Rent']]
y = df['Healthcare']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 42)
```

```python
x_train.columns
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
lr.score(x_train,y_train)
```

```python
### Exporting Model 
import pickle

with open('Health_care_model.pkl','wb') as f:
    pickle.dump(lr,f)
```
